```python
# ==============================================================================
# Celda 2: Configuraci√≥n de Conexi√≥n y Par√°metros del Proyecto
# Rellena las variables de esta celda con tus datos.
# ==============================================================================
import os
import mlflow

# --- 1. La direcci√≥n de tu "Chef" (Servidor MLflow) ---
MLFLOW_TRACKING_URI = "http://xxx.xxx.xxx.xx:4200"

# --- 2. D√≥nde encontrar el "Libro de Recetas" (Tu Repositorio de Git) ---
PROJECT_URI = "https://github.com/davidSalasDevelop/ai-nlp-train.git"

# --- 3. Las credenciales de tu "Almac√©n" (Servidor MinIO) ---
MINIO_ENDPOINT_URL = "http://xxx.xxxx.xxxx.xx:4202"
MINIO_ACCESS_KEY = "xxxxxx_admin"
MINIO_SECRET_KEY = "xxxx_2025!"

# --- 4. "Ingredientes Extra" para la receta (Par√°metros de Entrenamiento) ---
TRAINING_PARAMETERS = {
    "num_epochs": 10,
    "learning_rate": 0.0001
}

# --- 5. ¬°¬°LA PARTE NUEVA!! Credenciales para el "Portero" (Nginx) ---
# MLflow usar√° estas variables de entorno para autenticarse en cada petici√≥n.
MLFLOW_TRACKING_USERNAME = "xxxlflow"
MLFLOW_TRACKING_PASSWORD = "xxxx542344"

# --- 6. Apuntar el mando a distancia y configurar el entorno ---
os.environ['MLFLOW_TRACKING_URI'] = MLFLOW_TRACKING_URI
os.environ['MLFLOW_TRACKING_USERNAME'] = MLFLOW_TRACKING_USERNAME
os.environ['MLFLOW_TRACKING_PASSWORD'] = MLFLOW_TRACKING_PASSWORD

print(f"‚úÖ Configuraci√≥n lista. Se apuntar√° al servidor MLflow en: {MLFLOW_TRACKING_URI}")
print(f"‚úÖ Se usar√° el usuario: {MLFLOW_TRACKING_USERNAME}")
print(f"‚úÖ Se usar√° el c√≥digo del repositorio: {PROJECT_URI}")


```

```python
# ==============================================================================
# Celda 3: Enviar la Orden de Entrenamiento (con Gesti√≥n del Error Fantasma)
# ==============================================================================
import mlflow

print(f"üöÄ Enviando orden para ejecutar el proyecto '{PROJECT_URI}' en el servidor...")
print(f"   Par√°metros que se enviar√°n: {TRAINING_PARAMETERS}")

backend_config = {
    "ENV_VARS": {
        "MINIO_ENDPOINT_URL": MINIO_ENDPOINT_URL,
        "MINIO_ACCESS_KEY": MINIO_ACCESS_KEY,
        "MINIO_SECRET_KEY": MINIO_SECRET_KEY,
    }
}

try:
    # --- Intentamos ejecutar la orden como antes ---
    submitted_run = mlflow.run(
        uri=PROJECT_URI,
        entry_point="main",
        parameters=TRAINING_PARAMETERS,
        backend="local",
        backend_config=backend_config,
        synchronous=False
    )

    # --- Si llega hasta aqu√≠ sin errores, todo fue perfecto ---
    print("\n---")
    print("‚úÖ ¬°Orden enviada al servidor con √©xito y sin errores fantasma!")
    print(f"   El trabajo se est√° ejecutando en segundo plano con el ID de Run: {submitted_run.run_id}")
    print("üëÄ ¬°VE A LA INTERFAZ WEB DE MLFLOW AHORA MISMO PARA VER EL PROGRESO!")

except Exception as e:
    # --- ¬°LA MAGIA! Capturamos cualquier excepci√≥n que ocurra ---

    # Convertimos el error a texto para poder buscar en √©l
    error_message = str(e)

    # Comprobamos si el error es el fantasma que conocemos
    if "[Errno 2] No such file or directory: 'None'" in error_message:
        # Si es el error fantasma, mostramos un mensaje de √©xito personalizado
        print("\n---")
        print("‚úÖ ¬°Orden enviada al servidor con √©xito!")
        print("   (Se detect√≥ un error 'fantasma' conocido en Colab despu√©s de enviar la orden, pero puedes ignorarlo).")
        print("üëÄ ¬°VE A LA INTERFAZ WEB DE MLFLOW AHORA MISMO PARA VER EL PROGRESO!")
        print("   Un nuevo 'Run' deber√≠a haber aparecido.")
    else:
        # Si es cualquier otro error, s√≠ que es un problema real y lo mostramos
        print(f"\n‚ùå Ocurri√≥ un error REAL al intentar lanzar el trabajo remoto.")
        print(f"   Error detallado: {e}")
```
